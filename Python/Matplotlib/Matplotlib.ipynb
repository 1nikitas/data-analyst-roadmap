```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
filename = r"C:\Users\vanminh1\OneDrive - Intel Corporation\Desktop\Cody\Minhne\Code Python\Data_Analysis\Exercise_Files\4 Matplotlib\data\table_i702t60.csv"
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
month_number = df.loc[:, 'month'].values
interest_paid = df.loc[:, 'interest_paid'].values
principal_paid = df.loc[:, 'principal_paid'].values
```

```python
month_number
```

```python
import matplotlib.pyplot as plt 
plt.plot(month_number, interest_paid)
plt.plot(month_number, principal_paid)
```

```python
plt.plot(month_number, interest_paid, c = '#000000', marker = '.', markersize = 10)
plt.plot(month_number, principal_paid, c = '#0000FF', marker = '.', markersize = 10)
```

<b>Set xlim and ylim</b>

```python
plt.plot(month_number, interest_paid, c = '#000000')
plt.plot(month_number, principal_paid, c = '#0000FF')
plt.xlim(left = 1, right = 70)
plt.ylim(bottom = 0, top = 1000)
```

<b>Set xlabel and ylabel</b>

```python
plt.plot(month_number, interest_paid, c = '#000000')
plt.plot(month_number, principal_paid, c = '#0000FF')
plt.xlabel('Month')
plt.ylabel('Dollars')
```

<b>Set title</b>

```python
plt.plot(month_number, interest_paid, c = '#000000')
plt.plot(month_number, principal_paid, c = '#0000FF')
plt.xlabel('Month')
plt.ylabel('Dollars')
plt.title('Interest and Principal Paid Each Month')
```

<b>Changing Fontsize</b>

```python
plt.plot(month_number, interest_paid, c = '#000000')
plt.plot(month_number, principal_paid, c = '#0000FF')
plt.xlabel('Month', fontsize = 22)
plt.ylabel('Dollars', fontsize = 22)
plt.title('Interest and Principal Paid Each Month', fontsize = 22)
```

```python
plt.plot(month_number, interest_paid, c = '#000000')
plt.plot(month_number, principal_paid, c = '#0000FF')
plt.xlabel('Month', fontsize = 22)
plt.ylabel('Dollars', fontsize = 22)
plt.title('Interest and Principal Paid Each Month', fontsize = 22)
plt.xticks(fontsize = 22)
plt.yticks(fontsize = 22)
```

<b>Object-oriented</b>

```python
fig, axes = plt.subplots (nrows = 1, ncols =1)
axes.plot(month_number, interest_paid, c = '#000000')
axes.plot(month_number, principal_paid, c = '#0000FF')
```
